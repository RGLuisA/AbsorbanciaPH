# Graficas para los datos experimentales

## Agua (primera toma)

```python
import numpy as np
import matplotlib.pyplot as plt

# Cargar los datos del archivo
data = np.genfromtxt('C:\\Users\\rglui\\Downloads\\E5_agua 1.txt', delimiter='\t', skip_header=2)


# Separar los datos en longitud de onda y absorbancia
wavelength = data[:, 0]
absorbance = data[:, 1]

# Encontrar el máximo de absorción
max_abs = np.max(absorbance)
max_wavelength = wavelength[np.argmax(absorbance)]

# Definir los límites de longitud de onda para el espectro visible
# (Rojo: 620-750 nm, Naranja: 590-620 nm, Amarillo: 570-590 nm, 
# Verde: 495-570 nm, Azul: 450-495 nm, Violeta: 380-450 nm)
if 620 <= max_wavelength <= 750:
    color = 'Rojo'
elif 590 <= max_wavelength < 620:
    color = 'Naranja'
elif 570 <= max_wavelength < 590:
    color = 'Amarillo'
elif 495 <= max_wavelength < 570:
    color = 'Verde'
elif 450 <= max_wavelength < 495:
    color = 'Azul'
elif 380 <= max_wavelength < 450:
    color = 'Violeta'
else:
    color = 'Fuera del espectro visible'

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.plot(wavelength, absorbance, label='Absorbancia', color='blue')
plt.scatter([max_wavelength], [max_abs], color='red', label=f'Máximo: {max_wavelength} nm', zorder=5)
plt.title('Longitud de onda vs Absorbancia')
plt.xlabel('Longitud de onda (nm)')
plt.ylabel('Absorbancia')
plt.legend()
plt.grid(True)
plt.show()

max_wavelength, max_abs, color
```

 Dado que los primeros datos provienen de una saturación debido al propio funcionamiento del espectrofotometro, tenemos que es el segundo pico el que determina el máximo de absorción y no el primero, el cual se encuentra en 480nm con una absorbancia de .050 es decir, que se encuentra en la región del visible mas concretamente en el azul
